```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]\
```

```python
#combine all files in the list
#combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
data = []
for csv in all_filenames:
    frame = pd.read_csv(csv)
    print(os.path.basename(csv))
    frame['Year'] = os.path.basename(csv)
    data.append(frame)
#export to csv
#combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
bigframe = pd.concat(data, ignore_index=True) #dont want pandas to try an align row indexes
bigframe.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
```

```python
finalList = []
finalDict = {}
grouped = df.groupby(['Year'])
for key, value in grouped:
    dictionary = {}
    j = grouped.get_group(key).reset_index(drop=True)
    dictionary['Year'] = int(j.at[0, 'Year'])
    dictList = []
    anotherDict = {}
    
    for i in j.index:
        anotherDict['Player'] = j.at[i, 'Player']
        anotherDict['Team'] = j.at[i, 'Tm']
        anotherDict['GamesPlayed'] = int(j.at[i, 'G'])
        anotherDict['FG'] = int(j.at[i, 'FG'])
        anotherDict['FGP'] = j.at[i, 'FG%']
        anotherDict['Pos'] = j.at[i, 'Pos']
        dictList.append(anotherDict.copy())
    dictionary['Players'] = dictList
    finalList.append(dictionary)

import json
json.dumps(finalList)
```

```python
with open('data.json', 'w') as f:
    json.dump(finalList, f)
```
